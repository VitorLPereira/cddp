```python
Lspam=0
#total_de_emails=len(dados)
for i in range(len(email_treino)):
    if spam[i]=="spam":
        Lspam+=1

PSPAM=(Lspam/len(email_treino))
print("a probabilidade de um email ser spam é {0}".format(PSPAM))
```

```python
Lham=0
for i in range(len(email_treino)):
    if spam[i]=='ham':
        Lham+=1

PHAM=(Lham/len(email_treino))
print("a probabilidade de um email ser ham é {0}".format(PHAM))
```

```python
alospam=[]
aloham=[]
for i in range(len(email_treino)):
    j=(email_treino.values[i].split())
    #print(j)
    for k in j:
        if spam[i]=='spam':
            alospam.append(k)
        else:
            aloham.append(k)
            

(pd.Series(alospam).value_counts())/len(alospam)
```

probabilidade de cada palavra (word) acontecer se a mensagem na base treinamento é considerada SPAM

```python
(pd.Series(aloham).value_counts())/len(aloham)
```

A probabilidade de cada palavra acontecer se a mensagem na base treinamento é considerada HAM.

```python
valores_spam=pd.Series(alospam).value_counts()/len(alospam)
valores_ham=pd.Series(aloham).value_counts()/len(aloham)
Pspam=1
Pham=1
for i in range(len(email_treino)):
    j=(email_treino.values[i].split())
    for k in range (len(j)):
        Pspam*=((valores_spam[k])+(1/len(valores_spam)))
        Pham*=((valores_ham[k])+(1/len(valores_spam)))
    if Pspam>Pham:
        print('A probabilidade de ser spam é de {0} e a probabilidade de ser ham é de {1}, logo podemos concluir q o email é spam'.format(Pspam,Pham))
    else:
        print('A probabilidade de ser spam é de {0} e a probabilidade de ser ham é de {1}, logo podemos concluir q o email é ham'.format(Pspam,Pham))
```

Classificador:
    A probabilidade de um email ser spam ou não deve ser calculada pelo teorema de Bayes,
P(A | B) ={P(B | A) x P(A)} / {P(B)}, ou seja, probabilidade de um email ser spam dado um email = probabilidade de aparecer esse email, dado que é spam, vezes a probabilidade de ser spam. Tudo isso dividido pela probabilidade desse email aparecer. 
Porém no nosso classificador, será considerado spam ou ham que tiver a maior probabilidade calculada, por isso, por estarmos apenas comparando duas probabilidades, podemos retirar o divisor, sem alterar a conclusão de se o email é spam ou ham.

Feito isso temos que considerar também uma condicional no nosso cálculo de probabilidades, pode ser que apareçam na etapa teste palavras que não haviam aparecido anteriormente na etapa de treinamento. Fazemos por isso um acrescimo no calculo da probabilidade, 1 unidade dividida sobre o total de elementos, assim, se evita que a probabilidade seja zerada, e por ser apenas uma unidade no total de elementos, a diferença no resultado é mínima.

Chegamos então finalmente ao cálculo da probabilidade de um email, ser spam ou ham. Esse cálculo se dá pela multiplicação de cada palavra dentro do email, e a probabilidade de cada palavra pode ser calculado pela soma da frequencia absoluta dessa palavra dentre todas as palavras, somado a uma unidade.

Tendo calculadas as probabilidades temos que fazer uma análise de o que signiifica cada uma delas. Partindo do principio que a conclusão se um email é ou não spam serã feita a partir de calculo de apenas duas probabilidades, podemos afirmar que quando a probabilidade de ser spam for maior, o email será considerado como spam, e o mesmo ocorre para ham.

Além disso por estarmos multiplicando diversas probabilidades de várias palavras, os valores numéricos tratados são cada vez mais proximos de zero. Isso nao significa que o classificador deixou de funcionar, pois as probabilidades ainda estao sendo calculadas, o problema é que o python não tem capacidade para calcular tais valores, por isso, a partir de certo instante as probabilidades são arredondadas para zero.

Consequentemente seria inválido analisar a eficiência do classificador, uma vez que as probabilidades nao foram devidamente calculadas pelo python. Logo, calcular  falsos positivos, positivos verdadeiros, falsos negativos e negativos verdadeiros não faria sentido algum. No entanto poderíamos predizer que a única medida que seria discrepante seria a de falsos negativos(mensagens marcadas como não SPAM mas são SPAM), pois o codigo define em casos contrários à probailidade do email ser spam ser maior que ser ham, o email deve ser definido como ham. Por isso todos os casos não calculados são definidos como ham. Assim todos (exceto 2) os emails que são em realidade spams, no nosso classificador foram julgados como ham.
